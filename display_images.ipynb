```python
# Plotting images
import os
import matplotlib.pyplot as plt
import PIL
%matplotlib inline


string = '1,50'
path = '/data/abhijnya/Classifier_guidance/runs/rebuttal_classifiers/0,1,0,0/vanilla_LC_CUSTOM_t999_ninv50_ngen50/test_images/50'
images = [os.path.join(path,i) for i in os.listdir(path)]

# images = sorted(images, key=lambda x: (int(x.split('/')[-1].split('_')[1]), int(x.split('/')[-1].split('_')[3])))
rows = 10
fig = plt.figure(figsize=(15, 10))

for num, x in enumerate(images[:100]):
    img = PIL.Image.open(x)
    plt.subplot(rows,15,num+1)
    plt.axis('off')
    plt.imshow(img)

# plt.title(string)
plt.subplots_adjust(wspace=0.0, hspace=0)
```
